```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python

df=spark.read.table("default.iris_date")
     
```

```python
df2=df.withColumn("CLASS_NAME",regexp_replace(col("CLASS_NAME"),"Iris-",""))
```

```python
display(df2)
```

```python
df2.write.format("delta").mode("overwrite").saveAsTable("default.iris_date_transformed")
```
